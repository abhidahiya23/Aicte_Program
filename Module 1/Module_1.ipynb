# Task 1.4

In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock.
  ( VWAP = sum(price*volume)/sum(volume) ) 
To know more about VWAP , visit - VWAP definition 
{Hint : Create a new dataframe column ‘Month’.
The values for this column can be derived from the ‘Date” column by using appropriate pandas functions.
Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year.
Finally, calculate the vwap value for each month (i.e. for each group created).

```python
def date_extraction(df,column):
    df['Month'] = df[column].apply(lambda x : x.month)
    df['Year'] = df[column].apply(lambda x : x.year)
    df['Total_Cost'] = df['Close Price']*df['Total Traded Quantity']
```

```python
date_extraction(data,'Date')
```

```python
data.head()
```

```python
li = []
def vmap(df):
    # here we could use turnover(instead of Total_cost) too because the diff in values is not that much 
    num = df.groupby(by=['Month','Year'])['Total_Cost'].sum()  
    den = df.groupby(by=['Month','Year'])['Total Traded Quantity'].sum()
    value = num/den
    li.append(value)
```

```python
vmap(data)
```

# Task 1.5

1.5 Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter.
Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
# func to calculate the avg price over the last N days of the stock price data where N is a user defined
def avg_price(data,n):
    return data['Close Price'][-n:].mean()
```

```python
avg_price(data,90)
```

```python
# func to calculate the profit/loss percentage over the last N days.
# Profit/Loss percentage bw N days is the % change between the closing prices of the 2 days.
def profit_loss(data,n):
    change = data['Close Price'].iloc[-n] - data['Close Price'].iloc[-1]
    per  = (change*100)/data['Close Price'].iloc[-n]
    return per
```

```python
profit_loss(data,10)
```

```python
#Calculate the avg price AND the profit/loss per over the course of last -
#1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.

print(avg_price(data,7),profit_loss(data,7))
print(avg_price(data,14),profit_loss(data,14))
print(avg_price(data,30),profit_loss(data,30))
print(avg_price(data,120),profit_loss(data,120))
print(avg_price(data,180),profit_loss(data,180))
print(avg_price(data,365),profit_loss(data,365))
```
