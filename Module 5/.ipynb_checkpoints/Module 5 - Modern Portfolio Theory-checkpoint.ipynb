### Query 5.1 
5.1 For your chosen stock, calculate the mean daily return and daily standard deviation of returns, and then just annualise them to get mean expected annual return and volatility of that single stock. **( annual mean = daily mean * 252 , annual stdev = daily stdev * sqrt(252) )**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
jet = pd.read_csv("../DataSet/Small_Cap/JETAIRWAYS.csv")
```

```python
jet.head()
```

```python
# per_change
jet['per_change'] = jet['Close Price'].pct_change()
```

```python
# daily and annual mean and std
daily_mean = jet['per_change']
daily_std = jet['per_change']

annual_mean = daily_mean*252 
annual_std = daily_std*np.sqrt(252)
```

```python
# Importing new stoccks data
hdfc = pd.read_csv('../DataSet/Large_Cap/HDFC.csv')
pnb = pd.read_csv('../DataSet/Mid_Cap/PNB.csv')
maruti = pd.read_csv('../DataSet/Large_Cap/MARUTI.csv')
nifty = pd.read_csv('../DataSet/Nifty50/Nifty50.csv')
```

```python
np.unique(maruti['Series'])
```

```python
# Arrange all rows by value in index Series
maruti = maruti.set_index('Series')
# Drop all the rows with IL in Series
maruti = maruti.drop(['BL'])
maruti.reset_index(inplace=True,drop=True)
```

```python
# Arrange all rows by value in index Series
hdfc = hdfc.set_index('Series')
# Drop all the rows with IL in Series
hdfc = hdfc.drop(['BL','W2'])
hdfc.reset_index(inplace=True,drop=True)
```

```python
# crude['per_change'] = crude['Change %'].str.rstrip('%').astype(float)
```

```python
# all stocks
stocks = pd.concat([hdfc['Close Price'],
                    pnb['Close Price'],
                    nifty['Close'],
                    jet['Close Price'],
                    maruti['Close Price']],axis = 1)
```

```python
stocks_ret = stocks.pct_change()
```

```python
np.random.seed(42)
num_ports = 6000
all_weights = np.zeros((num_ports, len(stocks.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    # Weights
    weights = np.array(np.random.random(5))
    weights = weights/np.sum(weights)
    
    # Save weights
    all_weights[x,:] = weights
    
    # Expected return
    ret_arr[x] = np.sum( (stocks_ret.mean() * weights * 252))
    
    # Expected volatility
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(stocks_ret.cov()*252, weights)))
    
    # Sharpe Ratio
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]
```

```python
max_sr_vol = ret_arr[sharpe_arr.argmax()]
max_sr_ret = vol_arr[sharpe_arr.argmax()]
```

```python
vol_arr.min()
```

```python
min_vol_ret = ret_arr[vol_arr.argmin()]
min_vol_vol = vol_arr[vol_arr.argmin()]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_sr_vol, max_sr_ret,c='red', s=100,marker = '*')
plt.scatter(min_vol_vol,min_vol_ret,s=100)
plt.show()
```
